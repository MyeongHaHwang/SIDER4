## label_mapping.tsv.gz

The file appears to have a strange encoding and/or superfluous tab separation.

```python
columns = [
    'generic_names',
    'brand_names',
    'stitch_mapping',
    'stitch_id_flat',
    'stitch_id_sterio',
    'label_url',
    'label_id',
]
pandas.read_table('download/label_mapping.tsv.gz', encoding='iso-8859-1', names=columns)
```

```python
# elements per line
with gzip.open('download/label_mapping.tsv.gz', 'rt', encoding='iso-8859-1') as read_file:
    reader = csv.reader(read_file, delimiter='\t')
    counter = collections.Counter(len(row) for row in reader)
counter
```
